```python
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.image as mpimg
```

# Implementation of Gaussian kernel

```python
"""
Input kernal size
Returns gaussian_kernel, sum of all numbers in gaussian_kernel
http://www.subsurfwiki.org/wiki/Gaussian_filter
"""

def make_gaussian_kernel(size):
    x, y = np.mgrid[-size:size+1, -size:size+1] # returns a dense multi-dimensional “meshgrid”
    g = np.exp(-((x**2/float(size))+(y**2/float(size)))) # sigma = 1
    return g, g.sum()

kernel_size = 3
gaussian_kernel, S = make_gaussian_kernel(kernel_size)

plt.imshow(gaussian_kernel, cmap=plt.get_cmap('jet'), interpolation='nearest') # cmap=plt.cm.gray
plt.colorbar()
plt.title("Gaussian kernel")
plt.show()
```

# Read Image

```python
"""
Reading the image
Input: image file name
Returning its pixels in list and its width and height
"""

def read_image(filename):
    image_in = mpimg.imread(filename) 
    width_in, height_in = image_in.shape
    return image_in, width_in, height_in

image_in, width_in, height_in = read_image("Cameraman.png")
#image_in, width_in, height_in = read_image("Lenna.jpg")
```

```python
"""
Prints the image
Input: image pixels in list
"""

def print_image(image, title):
    plt.imshow(image, cmap=plt.cm.gray)
    plt.title(title)
    #plt.colorbar()
    plt.show()

print_image(image_in, "Original Image")
```

# Save an image to a file

```python
"""
Save an image to a file
Input: filename, image pixels in list
"""

def save_image(filename, image):
    mpimg.imsave(filename+".png", image, cmap=plt.cm.gray)
```

# Zero padding border policy

```python
"""
Fill the corners with 0's based on kernel_size
Input: image pixels in list, kernel_size
Returns the zero filled image pixels in list
"""

def zero_padding(image, kernel_size):
    width = image.shape[0]
    mat_zero = np.zeros(width)
    image_in_list = list(image)
    # making top and bottom rows zeros "kernel_size" times
    for i in range(kernel_size):
        image_in_list.insert(0, mat_zero)
        image_in_list.insert(len(image_in_list), mat_zero)
    # making right and left columns zeros "kernel_size" times
    for i in range(kernel_size):
        for k in range(len(image_in_list)):
            image_in_list[k] = list(image_in_list[k])
            image_in_list[k].insert(0, 0)
            image_in_list[k].insert(len(image_in_list[k]), 0)
    image_in_zero_pad = np.array(image_in_list)
    #width, height = image_in_zero_pad.shape
    return(image_in_zero_pad)

image_in_zero_pad = zero_padding(image_in, kernel_size)
print_image(image_in_zero_pad, "Zero padding border policy")
save_image("1.ZeroPadding", image_in_zero_pad)
```

# Blurring the image using Gaussian kernel

```python
"""
Blurring the image using Gaussian kernel
Input: zero padded image, kernel_size, gaussian_kernel, sum of all numbers in gaussian_kernel
Returns blurred image
"""

def blurrGaussian(image_in_zero_pad, kernel_size, gaussian_kernel, S):
    width, height = image_in_zero_pad.shape
    width_in = width-(2*kernel_size)
    height_in = height-(2*kernel_size)
    
    blurr_img = np.zeros((width_in, height_in))
    blurr_y = 0
    for x in range(0, width-(2*kernel_size)):
        blurr_x = 0
        for y in range(0, height-(2*kernel_size)):
            img_kernel = image_in_zero_pad[np.arange(x, 2*kernel_size+1+x), np.arange(y, 2*kernel_size+1+y)[:,None]]
            mul = np.multiply(img_kernel, gaussian_kernel)
            blurr_img[blurr_y][blurr_x] = mul.sum()/S
            blurr_x += 1
        blurr_y += 1
    return(blurr_img)
    
blurr_img = blurrGaussian(image_in_zero_pad, kernel_size, gaussian_kernel, S)
print_image(blurr_img, "Blurring the image using Gaussian kernel")
save_image("2.GaussianBlurr", blurr_img)
```

# Sub-sample an image

```python
"""
Reduce the image into half of its original size
Input: image pixels in list
"""

def subSample(image):
    width, height = image.shape
    subSam_img = np.zeros((width/2, height/2))
    
    subSam_img_y = 0
    for y in range(0, height, 2):
        subSam_img_x = 0
        for x in range(0, width, 2):
            subSam_img[subSam_img_x][subSam_img_y] = image[x][y]
            subSam_img_x += 1
        subSam_img_y += 1
    return(subSam_img)

subSam_img = subSample(image_in)
print_image(subSam_img, "Sub-sampled image")
save_image("3.SubSample", subSam_img)
```

# Image expansion

## Spreading the pixels

```python
"""
Spread the pixels in a pattern
Input: image pixels in list
Returns expanded image
"""

def spreadImg(image):
    width, height = image.shape
    subSam_exp_img = np.zeros((width*2, height*2))
    subSam_exp_img_y = 0

    for y in range(0, height):
        subSam_exp_img_x = 0
        for x in range(0, width):
            subSam_exp_img[subSam_exp_img_x][subSam_exp_img_y] = image[x][y]
            subSam_exp_img_x += 2
        subSam_exp_img_y += 2
    return(subSam_exp_img)

imageExp = spreadImg(subSam_img)
print_image(imageExp, "Spreading the pixels")
save_image("4.Expansion", imageExp)
```

## Bilinear interpolation

```python
"""
Filles the blank dots by Bilinear interpolation in an image
Input: expanded image
https://en.wikipedia.org/wiki/Bilinear_interpolation#Application_in_image_processing
"""

def bilinearInterpolation(subSam_exp_img):
    width, height = subSam_exp_img.shape
    
    """
    Filling the rows
    """
    for y in range(0, height-1, 2):
        # if the image width is even number
        if (width % 2 == 0):
            for x in range(0, width-1, 2):
                if (x == width-2):
                    subSam_exp_img[x+1][y] = subSam_exp_img[x][y]
                else:
                    if (y == height-2):
                        subSam_exp_img[x+1][y] = subSam_exp_img[x][y]
                    else:
                        a = ( ( (x+2)-(x+1) )/ float((x+2)-(x)) )*(subSam_exp_img[x][y])
                        b = ( ( (x+1)-(x)   )/ float((x+2)-(x)) )*(subSam_exp_img[x+2][y])
                        subSam_exp_img[x+1][y] = a + b
        else:
            for x in range(0, width-1, 2):
                if (y == height-2):
                        subSam_exp_img[x+1][y] = subSam_exp_img[x][y]
                else:
                    a = ( ( (x+2)-(x+1) )/ float((x+2)-(x)) )*(subSam_exp_img[x][y])
                    b = ( ( (x+1)-(x)   )/ float((x+2)-(x)) )*(subSam_exp_img[x+2][y])
                    subSam_exp_img[x+1][y] = a + b
    #print_image(subSam_exp_img, "Bilinear interpolation - rows")
    
    """
    Filling the columns
    """

    for y in range(0, height-1, 2):
        # if the image width is even number
        if (width % 2 == 0):
            for x in range(0, width-1, 2):
                if (x == width-2):
                    subSam_exp_img[x][y+1] = subSam_exp_img[x][y]
                else:
                    # last point in column
                    if (y == height-2):
                        subSam_exp_img[x][y+1] = subSam_exp_img[x][y]
                    else:
                        a = ( ( (y+2)-(y+1) )/ float((y+2)-(y)) )*(subSam_exp_img[x][y])
                        b = ( ( (y+1)-(y)   )/ float((y+2)-(y)) )*(subSam_exp_img[x][y+2])
                        subSam_exp_img[x][y+1] = a + b
        else:
            for x in range(0, width-1, 2):
                # last point in column
                if (y == height-2):
                    subSam_exp_img[x][y+1] = subSam_exp_img[x][y]
                else:
                    a = ( ( (y+2)-(y+1) )/ float((y+2)-(y)) )*(subSam_exp_img[x][y])
                    b = ( ( (y+1)-(y)   )/ float((y+2)-(y)) )*(subSam_exp_img[x][y+2])
                    subSam_exp_img[x][y+1] = a + b      
    #print_image(subSam_exp_img, "Bilinear interpolation - columns")
    
    """
    Filling the rows for center
    """

    for y in range(1, height-1, 2):
        # if the image width is even number
        if (width % 2 == 0):
            for x in range(0, width-1, 2):
                if (x == width-2):
                    subSam_exp_img[x+1][y] = subSam_exp_img[x][y]
                else:
                   # last point in column
                    if (y == height-2):
                        subSam_exp_img[x+1][y] = subSam_exp_img[x][y]
                    else:
                        a = ( ( (x+2)-(x+1) )/ float((x+2)-(x)) )*(subSam_exp_img[x][y])
                        b = ( ( (x+1)-(x)   )/ float((x+2)-(x)) )*(subSam_exp_img[x+2][y])
                        subSam_exp_img[x+1][y] = a + b
        else:
            for x in range(0, width-1, 2):
                if (y == height-2):
                    subSam_exp_img[x+1][y] = subSam_exp_img[x][y]
                else:
                    subSam_exp_img[x+1][y] = ( ( (x+2)-(x+1) )/ ((x+2)-(x)) )*subSam_exp_img[x][y] + ( ((x+1)-(x))/((x+2)-(x)) )*subSam_exp_img[x+2][y]
    #print_image(subSam_exp_img, "Bilinear interpolation - centers")
    
    """
    Filling last corner incomplete dots
    """

    for x in range(1, width-1, 2):
        subSam_exp_img[x][height-1] = subSam_exp_img[x-1][height-1]
    subSam_exp_img[width-1][height-1] = subSam_exp_img[width-2][height-1]
    return(subSam_exp_img)

subSam_exp_img = bilinearInterpolation(imageExp)
print_image(subSam_exp_img, "Bilinear interpolation")
save_image("5.BilinearInterpolation", subSam_exp_img)
```

```python
"""
Substract two images
Input: image pixels in list, image pixels in list
"""

def laplacianImage(image, blurr_img):
    laplacian_image = np.subtract(image, blurr_img)
    return(laplacian_image)
```

# Gaussian Pyramid

```python
"""
Creates Gaussian Pyramid of given levels
Input: image pixels in list, levels of pyramid
"""

def gaussianPyramid(image, l):
    laplacian_images = []
    for i in range(l):
        image_in_zero_pad = zero_padding(image, kernel_size)
        blurr_img = blurrGaussian(image_in_zero_pad, kernel_size, gaussian_kernel, S)
        lap_img = laplacianImage(image, blurr_img)
        laplacian_images.append(lap_img)
        image = subSample(blurr_img)
        print_image(image, "Gaussian Pyramid - level: "+str(i))
        save_image("6.GaussianPyramid"+str(i), image)
    return(image, laplacian_images)

level = 4
image_small, laplacian_images = gaussianPyramid(image_in, level)
```

```python
"""
Printing Laplacian images
"""

for img in laplacian_images:
    print_image(img, "Laplacian images")
```

# Laplacian Pyramid

```python
"""
Creates Laplacian Pyramid of given levels
Input: image pixels in list, levels of pyramid
"""

def laplacianPyramid(small_img, laplacian_images, l):
    for i in range(l):
        #print_image(small_img)
        imageExp = spreadImg(small_img)
        #print_image(imageExp)
        small_img = bilinearInterpolation(imageExp)
        #print_image(small_img)
        
        small_img = np.add(small_img, laplacian_images[l-i-1])
        
        save_image("7.LaplacianPyramid"+str(i), small_img)
        #small_img = mpimg.imread("LaplacianPyramid"+str(i)+".png")
        #small_img = small_img[:, :, 0]
        print_image(small_img, "Laplacian Pyramid - level: "+str(i))
        #print("laplacianPyramid", image_in.shape)
        #width_in, height_in = image_in.shape
        #small_img, width, height = read_image("LaplacianPyramid"+str(i)+".jpg")
    return(small_img)
    
image_big = laplacianPyramid(image_small, laplacian_images, level)
```
