# Translation

```python
# moving x by 50, and y by 50
trans_mat = [[1, 0, 50], [0, 1, 50], [0, 0, 1]]
print(trans_mat)
print(np.array(trans_mat).shape)
trans_mat = np.array(trans_mat)
```

```python
s = (500, 500)
mat_zero = np.zeros(s)
print(mat_zero)
```

```python
x, y = image_in.shape

image_new = []

for x_val in range(int(x)):
    for y_val in range(int(y)):
        mat = np.dot(trans_mat, np.array([x_val, y_val, 1]))
        #print(mat)
        try:
            mat_zero[mat[0]][mat[1]] = image_in[x_val][y_val]
        except:
            pass
```

```python
mat_zero
```

```python
plt.imshow(mat_zero, cmap=plt.cm.gray)
plt.show()
print(mat_zero.shape)
```

```python
plt.close()
```

# Rotation

```python
# Rotating by 20 degrees
a = 20
a= np.deg2rad(a)
trans_mat = [[np.cos(a), -np.sin(a), 0], [np.sin(a), np.cos(a), 0], [0, 0, 1]]
print(trans_mat)
print(np.array(trans_mat).shape)
trans_mat = np.array(trans_mat)
```

```python
s = (500, 500)
mat_zero = np.zeros(s)
mat_zero = np.ones(s)
print(mat_zero)
```

```python
x, y = image_in.shape

image_new = []

for x_val in range(int(x)):
    for y_val in range(int(y)):
        mat = np.dot(trans_mat, np.array([x_val, y_val, 1]))
        #print(mat)
        mat_zero[int(abs(mat[0]))][int(abs(mat[1]))] = image_in[x_val][y_val]
#         try:
#             mat_zero[mat[0]][mat[1]] = image_np[x_val][y_val]
#         except:
#             pass
```
